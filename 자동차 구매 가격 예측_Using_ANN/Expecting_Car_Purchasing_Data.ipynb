```python
import tensorflow.keras
#ìˆœì°¨ í˜•íƒœë¡œ ë„¤íŠ¸ì›Œí¬ êµ¬ì¶•
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
#inputì´ 5ê°œì˜ ì—´(ë°ì´í„°) -> 5ê°œì˜ ë‰´ëŸ°ì´ ì²« ì¸µì— ì¡´ìž¬
#ì²« ë²ˆì§¸ hidden layer
model.add(Dense(40, input_dim = 5, activation = 'relu'))
#ë‘ ë²ˆì§¸ ì€ë‹‰ì¸µ
model.add(Dense(40, activation = 'relu'))
#ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ëŠ” ì¸µ ->íŠ¹ì • ê°’ì„ ì˜ˆì¸¡í•´ì•¼ í•˜ë¯€ë¡œ linearì‚¬ìš©
model.add(Dense(1, activation = 'linear'))
```

```python
model.summary()
```

```python
#optimizerëŠ” adam optimizer, loss functionì€ ìµœì†Œì œê³±ë²•
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
#epochëŠ” forward+backwardë¥¼ í•©ì¹œ í•œ ë²ˆì˜ ê³¼ì •, epochsëŠ” í•™ìŠµíšŸìˆ˜
#batch size ëŠ” í•œ ë²ˆì˜ ë„¤íŠ¸ì›Œí¬ì— ëª‡ ê°œì˜ ë°ì´í„°ë¥¼ ë„£ì„ ì§€ ê³ ë ¤í•¨
#batchê°€ ì ìœ¼ë©´ ë°˜ë³µíšŸìˆ˜ê°€ ë§Žì•„ì§, batchê°€ ì»¤ì§€ë©´ ëŠë ¤ì§
#batchë¥¼ 25->50ìœ¼ë¡œ ëŠ˜ë ¸ì„ ë•Œ lossê°€ ëŠ˜ì—ˆë‹¤. ì ì ˆí•œ batch sizeë¥¼ ì°¾ëŠ” ê²ƒì´ ì¤‘ìš”
# dataë¥¼ ë‹¤ ì‚¬ìš©í•˜ì§€ ì•Šì„ ê²½ìš° ë‹¤ìŒ í•™ìŠµë•Œ ê·¸ ë¶€ë¶„ë¶€í„° ì‚¬ìš©í•œë‹¤.
# ë³€ìˆ˜ë¥¼ ë‹¤ë¥´ê²Œí•´ì„œ ë‘ ë°©ë²•ì„ ë¹„êµí•˜ê³  ì‹¶ìœ¼ë©´ ë‹¤ì‹œ ì²˜ìŒë¶€í„° ëŒë¦¬ìž.
# verboseëŠ” ë°‘ì— loss, val_lossë“± ë‹¤ì–‘í•œ ì •ë³´ë¥¼ í‘œì‹œ, (0,1,2)
# validation_splitì€ training dataë¥¼ ì¼ì • ë¹„ìœ¨ë§Œí¼ ë‚˜ëˆ„ì–´ ê²€ì¦ì— ì‚¬ìš©í•¨
# val_lossëŠ” ê·¸ ë°ì´í„°ë“¤ì˜ loss
# overfitting ë°©ì§€
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50, verbose=1, validation_split = 0.2)
```
