```python
import tensorflow.keras
#순차 형태로 네트워크 구축
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
#input이 5개의 열(데이터) -> 5개의 뉴런이 첫 층에 존재
#첫 번째 hidden layer
model.add(Dense(40, input_dim = 5, activation = 'relu'))
#두 번째 은닉층
model.add(Dense(40, activation = 'relu'))
#결과를 출력하는 층 ->특정 값을 예측해야 하므로 linear사용
model.add(Dense(1, activation = 'linear'))
```

```python
model.summary()
```

```python
#optimizer는 adam optimizer, loss function은 최소제곱법
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
#epoch는 forward+backward를 합친 한 번의 과정, epochs는 학습횟수
#batch size 는 한 번의 네트워크에 몇 개의 데이터를 넣을 지 고려함
#batch가 적으면 반복횟수가 많아짐, batch가 커지면 느려짐
#batch를 25->50으로 늘렸을 때 loss가 늘었다. 적절한 batch size를 찾는 것이 중요
# data를 다 사용하지 않을 경우 다음 학습때 그 부분부터 사용한다.
# 변수를 다르게해서 두 방법을 비교하고 싶으면 다시 처음부터 돌리자.
# verbose는 밑에 loss, val_loss등 다양한 정보를 표시, (0,1,2)
# validation_split은 training data를 일정 비율만큼 나누어 검증에 사용함
# val_loss는 그 데이터들의 loss
# overfitting 방지
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50, verbose=1, validation_split = 0.2)
```

```python
epochs_hist
```
